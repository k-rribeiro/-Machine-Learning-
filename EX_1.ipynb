```python
#Trecho do código citado na questão retirado do notebook-poo ---------------------------
def create_matrix(n_rows, m_cols):
    matrix = np.zeros((n_rows, m_cols)) # Initialize with correct dimensions
    for i in range(n_rows):
        matrix[i,:] = i # Set all elements in row i to i
    return matrix

# Exemplo de uso da função:
n_rows = 10 # Número de linhas
m_cols = 4 # Número de colunas

print('Matriz X:')
X = create_matrix(n_rows, m_cols)
# Exibindo a matriz
for row in X:
    print(row)

print('Vetor y:')
y = create_matrix(n_rows, 1)
# Exibindo a matriz
for row in y:
    print(row)
#------------------------------------------------------------------------------



#codigo alterado para mostrar se de fato funciona a classe de separar o treino do teste
#ex: 0.4 é 40% do total
altera = separa_treino_e_teste(tam_treino=0.4)
treino_x, teste_x, treino_y, teste_y = altera.corte(X, y)

#aqui irá exibir os treinos
print("--- exibir treinos ---")
for row_x, row_y in zip(treino_x, treino_y):
    print(f"X: {row_x} | y: {row_y}")

#aqui irá exibir os testes
print("--- exibir testes ---")
for row_x, row_y in zip(teste_x, teste_y):
    print(f"X: {row_x} | y: {row_y}")
```

```python
#Abaixo se encontra o código referenta a questão 1----------------------------
class separa_treino_e_teste:

    def __init__(self, tam_treino):
        self.tam_treino = tam_treino

    def corte(self, X, y):
        #aqui calcula o indice onde o corte vai ser feito
        corte_x = int(len(X) * self.tam_treino)

        #optei por usar o fatiamento para separar os dados de treino e teste
        treino_x = X[:corte_x]
        teste_x = X[corte_x:]
        treino_y = y[:corte_x]
        teste_y = y[corte_x:]

        return treino_x, teste_x, treino_y, teste_y
```
